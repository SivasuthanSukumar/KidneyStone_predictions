```python
x= df.drop(['id', 'target'], axis= 1)
```

```python
y= df['target']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,  x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state= 101)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

**model building**

*Logistic Regression*

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred_lr= lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm= confusion_matrix(y_test, y_pred_lr) 
cm
```

```python
cr= classification_report(y_test, y_pred_lr)
print(cr)
```

*Decision Trees*

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred_tree= dt.predict(x_test)
```

```python
cm= confusion_matrix(y_pred_tree, y_test)
cm
```

```python
cr= classification_report(y_pred_tree, y_test)
print(cr)
```

*Random Forest*

```python
# Import the RandomForestClassifier from the sklearn.ensemble module
from sklearn.ensemble import RandomForestClassifier

# Create a RandomForestClassifier model with 20 decision trees
rf = RandomForestClassifier(n_estimators=20)
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred_rf= rf.predict(x_test)
```

```python
cm= confusion_matrix(y_pred_rf, y_test)
cm
```

```python
cr= classification_report(y_pred_rf, y_test)
print(cr)
                          
```

*KNN*

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
y_predict_knn= knn.predict(x_test)
```

```python
cm= confusion_matrix(y_predict_knn, y_test)
cm
```

```python
cr= classification_report(y_predict_knn, y_test)
print(cr)
```

best model: Random Forest Classifier
